```python
expenses = []

def add_expense():
    name = input("Enter your Expense name: ")
    amount = float(input("Add Expense : "))
    category = input("Enter expense category: ")

    expense = {
        "name": name,
        "amount": amount,
        "category": category
    }

    expenses.append(expense)

    print(f"Expense '{name}' added successfully")


def view_expenses():
    if not expenses:
        print("No expenses to display")
    else:
        print("\n---All Expenses---")

        for expense in expenses:
            print(
                f"Name: {expense['name']}, "
                f"Amount: ${expense['amount']}, "
                f"Category: {expense['category']}"
            )


def view_expenses_by_category():
    category_to_view = input("Enter expense category to view: ")

    found_in_category = False

    print(f"\n--- Expenses in Category: {category_to_view} ---")

    for expense in expenses:
        if expense['category'].lower() == category_to_view.lower():

            print(
                f"Name: {expense['name']}, "
                f"Amount: ${expense['amount']}, "
                f"Category: {expense['category']}"
            )

            found_in_category = True

    if not found_in_category:
        print(f"No expenses found in category '{category_to_view}'")


def show_total_expenses():

    search_category = input("Enter category: ")

    total = 0

    for expense in expenses:

        if expense["category"].lower() == search_category.lower():

            total += expense["amount"]

    print(total)


while True:

    print("\n==== Expense Tracker ====")
    print("1. Add Expense")
    print("2. View Expense")
    print("3. View Expenses by Category")
    print("4. Show Total Expenses")
    print("5. Exit")

    choice = input("Enter your choice (1/2/3/4/5): ")

    if choice == "1":
        add_expense()

    elif choice == "2":
        view_expenses()

    elif choice == "3":
        view_expenses_by_category()

    elif choice == "4":
        show_total_expenses()

    elif choice == "5":
        print("Exiting Expense Tracker")
        break

    else:
        print("Invalid choice. Please enter a number between 1 and 5.")
```
